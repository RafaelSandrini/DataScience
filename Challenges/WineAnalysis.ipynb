```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('winequality-red.csv')
df.info()
```

```python
df.head()
```

```python
# features are attributes to use to get the target
# target is what we want to predict
```

```python
df['quality'] = df['quality'].map({6:'good',7:'good',8:'good',3:'bad',4:'bad',5:'bad'})
```

```python
df['quality'].unique()
```

```python
y = df['quality']
y
```

```python
X = df.drop(columns='quality')
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

# para deixar os dados na mesma escala
from sklearn.preprocessing import StandardScaler

# Sequencia de operacoes
from sklearn.pipeline import Pipeline
```

```python
etapas = [('scaler',StandardScaler()),
          ('logit',LogisticRegression())]

pile_logit = Pipeline(etapas).fit(X_train, y_train)

pile_logit
```

```python
y_pred = pile_logit.predict(X_test)
y_pred[:5]
```

```python
y_test[:5]
```

### Metricas de avaliacao

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
#passo 1 - construcao do modelo
etapas = [('scaler',StandardScaler()),
          ('logit',LogisticRegression())]

pile_logit = Pipeline(etapas).fit(X_train, y_train)


#passo 2 - avaliacao do modelo

y_pred = pile_logit.predict(X_test)

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.show()
print(classification_report(y_test, y_pred))
```
